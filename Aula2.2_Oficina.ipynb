```python
#Importe as bibliotecas pandas e matplotlib
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Dados em DataFrame
games = pd.read_csv('dadosoficina04.txt')
games.head()
```

```python
#Visualização que mostre a proporção de vendas global entre gêneros de jogos 1
generos = games.groupby(by='Genre').sum() 
generos
```

```python
#Visualização que mostre a proporção de vendas global entre gêneros de jogos 2
generos = generos.sort_values(by='Global_Sales',ascending=False)
porcentagens = generos['Global_Sales']*100.0/generos['Global_Sales'].sum()
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(generos.index,porcentagens)
plt.title('Proporção de vendas de jogos por gênero')
plt.show()
```

```python
#visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015 1
generos = games.groupby(by='Genre').sum() 
generos
```

```python
#visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015 2
generos = games.groupby(by='Year').sum() 
generos
```
