```python
# enunciado 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
# enunciado 2
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=mpg)
```

```python
# enunciado 3
plt.figure(figsize=(8,5))
plt.title('Relação entre acelaração, milhas por galão e modelo_ano')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=mpg)
plt.show()
```

```python
# Percebe-se, à partir da relação de dados do gráfico acima, que quanto mais novo o modelo, mais milhas por galão o automóvel faz
#Ou seja, quanto mais novo o automóvel, mais econômico.
```
