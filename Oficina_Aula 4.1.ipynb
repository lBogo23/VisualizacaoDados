```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribuição de acordo com a cidade')
sns.stripplot(x="fare", y="embark_town", data=titanic)
plt.show()
```

```python
sns.catplot(x="fare", y="embark_town", hue = 'sex', col = 'survived', kind = 'swarm', data=titanic)
```
