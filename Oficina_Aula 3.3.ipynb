```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('dadosoficina07.txt')
df.head()
```

```python
plt.figure(figsize=(20,5))
plt.title('vendas ao longo dos meses')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()
# A tendência é crescente. O numero de vendas aumenta ao longo dos meses.
```
